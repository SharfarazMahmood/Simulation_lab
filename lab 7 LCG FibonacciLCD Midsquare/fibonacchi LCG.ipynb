```
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```
seed1=2
seed2=3
z=[seed1,seed2]

m=32
u=[]

for i in range (2,102):
    zi = ( z[i-1] + ( 3*(z[i-2])**2 ) ) % m
    z.append(zi)
    #print(zi)
    u.append(zi/m)

print()
print(z)
print(u)

unique=set(z)
print(unique)
print(len(unique))
```
