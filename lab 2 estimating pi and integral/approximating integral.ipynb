```
import numpy as np
import math 

#limits
a=1
b=3

#data points
n=10000

randvalue=np.random.uniform(a,b, n)
print(randvalue)
```

```
#### f= 2x+3
f_sum=0
f_sqr_sum=0
for i in range(n):
    x=np.random.uniform(a,b)
    f_value= (2*x)+3
    f_sqr= f_value**2

    f_sum=f_sum + f_value
    f_sqr_sum = f_sqr_sum + f_sqr

f_avg= f_sum/n
f_sqr_avg=f_sqr_sum/n
#print(func_avg)

value_of_integral=(b-a)*f_avg
print(value_of_integral)

error=((b-a)/ math.sqrt(n) ) * math.sqrt(f_sqr_avg - f_avg**2) 
print(error)
```

```
##f = xe^x

a=0
b=2
n=1000
f_sum=0
f_sqr_sum=0
for i in range(n):
    x=np.random.uniform(a,b)
    f_value= x * math.exp(x)
    f_sqr= f_value**2

    f_sum=f_sum + f_value
    f_sqr_sum = f_sqr_sum + f_sqr

f_avg = f_sum/n
f_sqr_avg=f_sqr_sum/n
#print(func_avg)

value_of_integral=(b-a)*f_avg
print(value_of_integral)

error=((b-a)/ math.sqrt(n) ) * math.sqrt(f_sqr_avg - f_avg**2) 
print(error)
```
