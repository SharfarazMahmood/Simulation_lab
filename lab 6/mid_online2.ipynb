```
import numpy as np
import matplotlib.pyplot as plt
import math


np.random.seed(1)

#limits
a=0
b=2
c=5

trials = [100,1000,5000]
x = ['100','1000','5000']

#randvalue=np.random.uniform(a,b, n)
#print(randvalue)
```

```
#### f1= sqrt(25-x^2)
error=[0,0,0]
errory=[0,0,0]
area=[0,0,0]
areay=[0,0,0]

totalarea=[0,0,0]

for i in range(3):
    f1_sum=0
    f1_sqr_sum=0
    f_sum=0
    f_sqr_sum=0
    for j in range(trials[i]):
        x=np.random.uniform(a,b)
        f1_value= 5*(x**2)
        f1_sqr= f1_value**2

        f1_sum=f1_sum + f1_value
        f1_sqr_sum = f1_sqr_sum + f1_sqr

        y=np.random.uniform(b,c)
        f_value= (-(2/3))*y+(5/3)
        f_sqr= f_value**2

        f_sum=f_sum + f_value
        f_sqr_sum = f_sqr_sum + f_sqr

    f1_avg= f1_sum/trials[i]
    f1_sqr_avg=f1_sqr_sum/trials[i]
    #print(func_avg)
    f_avg= f_sum/trials[i]
    f_sqr_avg=f_sqr_sum/trials[i]
    

    value_of_integral=(b-a)*f1_avg
    area[i]= value_of_integral
    #print(area[i])
    value_of_integral=(c-b)*f_avg
    areay[i]= value_of_integral
    

    error[i]=((b-a)/ math.sqrt(trials[i]) ) * math.sqrt(f1_sqr_avg - f1_avg**2) 
    errory[i]=((c-b)/ math.sqrt(trials[i]) ) * math.sqrt(f_sqr_avg - f_avg**2) 
    #print(error[i])
    totalarea[i]=(area[i]+areay[i])

print(area)
print(areay)
print()
print(totalarea)

print()
print(error)
print(errory)
```
