```
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```
x_a=[10]
y_a=[0]
x_b=[0]
y_b=[10]
x_c=[10]
y_c=[10]


s_a=3
s_b=5
s_c=7



dist_cb= math.sqrt((x_c[0]-x_b[0])**2 + (y_c[0]-y_b[0])**2)
dist_ac= math.sqrt((x_a[0]-x_c[0])**2 + (y_a[0]-y_c[0])**2)
dist_ba= math.sqrt((x_b[0]-x_a[0])**2 + (y_b[0]-y_a[0])**2)

d_cb= [0]
d_ac= [0]
d_ba= [0]

d_cb[0]= dist_cb
d_ac[0]= dist_ac
d_ba[0]= dist_ba

print(d_cb[0])
print(d_ac[0])
print(d_ba[0])
```

```

xa_n=0
ya_n=0
xb_n=0
yb_n=0
xc_n=0
yc_n=0

count_cb=0
count_ac=0
count_ba=0


for t in range(1,6):

    sin_b=(y_c[t-1] - y_b[t-1])/dist_cb
    cos_b=(x_c[t-1] - x_b[t-1])/dist_cb

    sin_c=(y_a[t-1] - y_c[t-1])/dist_ac
    cos_c=(x_a[t-1] - x_c[t-1])/dist_ac

    sin_a=(y_b[t-1] - y_a[t-1])/dist_ba
    cos_a=(x_b[t-1] - x_a[t-1])/dist_ba

    xc_n= x_c[t-1] + s_c* cos_c
    yc_n= y_c[t-1] + s_c* sin_c
    xb_n= x_b[t-1] + s_b* cos_b
    yb_n= y_b[t-1] + s_b* sin_b
    xa_n= x_a[t-1] + s_a* cos_a
    ya_n= y_a[t-1] + s_a* sin_a

    #print(str(xd_n)+" "+str(xc_n)+" "+str(xb_n)+" "+str(xa_n))
    #print(str(yd_n)+" "+str(yc_n)+" "+str(yb_n)+" "+str(ya_n))

    x_c.append(xc_n)
    y_c.append(yc_n)
    x_b.append(xb_n)
    y_b.append(yb_n)
    x_a.append(xa_n)
    y_a.append(ya_n)

    dist_cb= math.sqrt((x_c[t]-x_b[t])**2 + (y_c[t]-y_b[t])**2)
    dist_ba= math.sqrt((x_b[t]-x_a[t])**2 + (y_b[t]-y_a[t])**2)
    dist_ac= math.sqrt((x_a[t]-x_c[t])**2 + (y_a[t]-y_c[t])**2)

    d_cb.append(dist_cb)
    d_ba.append(dist_ba)
    d_ac.append(dist_ac)

    print(" "+str(dist_cb)+" "+str(dist_ac)+" "+str(dist_ba))

    if  dist_cb<8:
        count_cb=count_cb+1
        print("b_Fire_c "+ str(t))
    if  dist_ac<8:
        count_ac=count_ac+1
        print("c_Fire_a "+ str(t))
    if  dist_ba<8:
        count_ba=count_ba+1
        print("a_Fire_b "+ str(t))    


print()

print("b shoots c "+str(count_cb)+" times")
print("c shoots a "+str(count_ac)+" times")
print("a shoots b "+str(count_ba)+" times")

#print(x_d)
#print(y_d)   
#print()
#print(x_c)
#print(y_c)   
#print()
#print(x_b)
#print(y_b)   
#print()
#print(x_a)
#print(y_a)   
#print()

```
