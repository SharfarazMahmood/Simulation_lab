```
import matplotlib.pyplot as plt
# line 1 points 
x1 = [1,2,3,6,8,11] 
y1 = [2,4,1,3,8,5] 
# plotting the line 1 points  
plt.plot(x1, y1, label = "line 1", color='green', linestyle='dashed', linewidth = 1, 
         marker='o', markerfacecolor='blue', markersize=8 )
  
# line 2 points 
x2 = [1,2,3,6,8,11] 
y2 = [4,1,3,3,8,5] 
# plotting the line 2 points  
plt.plot(x2, y2, label = "line 2") 

#limit
plt.xlim(1,15)
plt.ylim(1,10)


# naming the x-axis 
plt.xlabel('x - axis') 
# naming the y-axis 
plt.ylabel('y - axis') 
# plot title 
plt.title('Graph') 



# labels for bars 
tick_label = ['1', '2', '3', '4', '5','6'] 
  
# plotting a bar chart 
plt.bar(x2, y2, tick_label = tick_label, width = 0.5, color = [ 'green']) 
```
