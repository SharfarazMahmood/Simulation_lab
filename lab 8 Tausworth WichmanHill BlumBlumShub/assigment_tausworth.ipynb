```
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```
trials=[400,4000,20000] #number of bits needed for 100,1000,5000 ran. numbers
num_of_rand_numbers=[100,1000,5000]

u_list=[]
r=3
q=5
l=4
a=2**l

values_list=[]
for w in range(3):
    b_i=[1,1,1,1,1]
    u=[]
    

    values=[]
    for i in range(5,trials[w]):
        new_b = ( (b_i[i-(r)]) + (b_i[i-(q)]) ) % 2
        b_i.append(new_b)
    #print(b_i)

    for i in range (0, trials[w]-1 ,l):  #since b_i has 0 to 'trails[i]-1' number of bits
        bit_segment=b_i[i:i+l]
        #print(bit_segment)

        strl=""
        for j in bit_segment:
            strl=strl+str(j)
        #print(strl)
        
        value= int (strl,2)  #converting base (binary to integer)
        values.append(value)
        #print(value)
        
        new_u=value/a
        u.append(new_u)
        #print(new_u)
    u_list.append(u)
    values_list.append(values)

print(values_list[0])
print(u_list[0])
print()
print(values_list[1])
print(u_list[1])
print()
print(values_list[2])
print(u_list[2])

print()
print()

print(set(values_list[0]))
print(len(set(values_list[0])))
print(set(values_list[1]))
print(len(set(values_list[1])))
print(set(values_list[2]))
print(len(set(values_list[2])))
```
