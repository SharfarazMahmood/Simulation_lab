```
n=100
z1=[5]
u1=[]
m1= 15

for i in range(1,n):
    new_z1= (z1[i-1]+3) % m1
    z1.append(new_z1)
    u1_new= new_z1/m1
    u1.append(u1_new)

print(z1)
print(u1)

z2=[7]
u2=[]
m2=16

for i in range(1,n):
    new_z2= (z2[i-1]**2 + 5) % m2
    z2.append(new_z2)
    u2_new= new_z2/m2
    u2.append(u2_new)

print(z2)
print(u2)

u = []
for i in range(0,n-1): #since previous two for loops iterate from 1 to 100
    u_new = u1[i]+u2[i]
    u.append(u_new-int(u_new))

print(u)
```
