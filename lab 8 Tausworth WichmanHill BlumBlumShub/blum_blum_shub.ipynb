```
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```
p=11
q=19
m=p*q

x=[3] #x[0]=3; given 
b_i=[]

for i in range(1,1000):
    new_x = (x[i-1]**2) % m
    x.append(new_x)

    if new_x%2 == 0:
        b_i.append(0)
    else:
        b_i.append(1)

#print(x)
print(b_i)
```

```


u=[]
l=4
a=2**l
vals=[]
for i in range(0,999,l):
    bit_segment  = b_i[i:i+l]
    #print(bit_segment)

    strl=""
    for j in bit_segment:
        strl=strl+str(j)
    #print(strl)
    value= int (strl,2)
    #print(value)
    u.append(value/a)
    vals.append(value)


print(vals)
print(u)
```
