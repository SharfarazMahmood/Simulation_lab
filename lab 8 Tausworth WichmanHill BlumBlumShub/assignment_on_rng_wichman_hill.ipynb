```
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```
trials=[102,1002,5002]
u_list=[]
u1_list=[]
u2_list=[]
u3_list=[]

z1_list=[]
z2_list=[]
z3_list=[]

for q in range(3):
    z1=[12,7]
    u1=[]
    m1= 16

    z2=[3,5]
    u2=[]
    m2=17

    z3=[2,7]
    u3=[]
    m3=15

    u = []
    for i in range(2,trials[q]):
        new_z1= ( (13*z1[i-1]) + (11*z1[i-2]) +3) % m1
        z1.append(new_z1)
        u1_new= new_z1/m1
        u1.append(u1_new)

        new_z2= ( (12*(z2[i-1]**2)) + (13*z2[i-2]) ) % m2
        z2.append(new_z2)
        u2_new= new_z2/m2
        u2.append(u2_new)

        new_z3= ( (z3[i-1]**3) + (z3[i-2]**2) ) % m3
        z3.append(new_z3)
        u3_new= new_z3/m3
        u3.append(u3_new)

        u_new = u1[i-2]+ u2[i-2]+ u3[i-2]
        u.append(u_new-int(u_new))
    z1_list.append(z1)
    u1_list.append(u1)
    z2_list.append(z2)
    u2_list.append(u2)
    z3_list.append(z3)
    u3_list.append(u3)

    u_list.append(u)


#print(z1_list)
#print(u1_list)
#print()
#print(z2_list)
#print(u2_list)
#print()
#print(z3_list)
#print(u3_list)
#print()

#print(u_list)


print(set(u_list[0]))
print(len(set(u_list[0])))
print(set(u_list[1]))
print(len(set(u_list[1])))
print(set(u_list[2]))
print(len(set(u_list[2])))
```
