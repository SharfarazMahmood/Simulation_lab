```
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```
x_a=[10]
y_a=[10]
x_b=[30]
y_b=[10]
x_c=[30]
y_c=[30]
x_d=[0]
y_d=[30]

s_a=10
s_b=15
s_c=12
s_d=5

dist_dc= math.sqrt((x_d[0]-x_c[0])**2 + (y_d[0]-y_c[0])**2)
dist_cb= math.sqrt((x_c[0]-x_b[0])**2 + (y_c[0]-y_b[0])**2)
dist_ba= math.sqrt((x_b[0]-x_a[0])**2 + (y_b[0]-y_a[0])**2)
dist_ad= math.sqrt((x_a[0]-x_d[0])**2 + (y_a[0]-y_d[0])**2)

print(dist_dc)
print(dist_cb)
print(dist_ba)
print(dist_ad)
```

```
xa_n=0
ya_n=0
xb_n=0
yb_n=0
xc_n=0
yc_n=0
xd_n=0
yd_n=0

count_dc=0
count_cb=0
count_ba=0
count_ad=0


for t in range(1,21):
    
    sin_c=(y_d[t-1] - y_c[t-1])/dist_dc
    cos_c=(x_d[t-1] - x_c[t-1])/dist_dc

    sin_b=(y_c[t-1] - y_b[t-1])/dist_cb
    cos_b=(x_c[t-1] - x_b[t-1])/dist_cb

    sin_a=(y_b[t-1] - y_a[t-1])/dist_ba
    cos_a=(x_b[t-1] - x_a[t-1])/dist_ba

    sin_d=(y_a[t-1] - y_d[t-1])/dist_ad
    cos_d=(x_a[t-1] - x_d[t-1])/dist_ad

    
    xc_n= x_c[t-1] + s_c* cos_c
    yc_n= y_c[t-1] + s_c* sin_c
    xb_n= x_b[t-1] + s_b* cos_b
    yb_n= y_b[t-1] + s_b* sin_b
    xa_n= x_a[t-1] + s_a* cos_a
    ya_n= y_a[t-1] + s_a* sin_a
    xd_n= x_d[t-1] + s_d* cos_d
    yd_n= y_d[t-1] + s_d* sin_d

    #print(str(xd_n)+" "+str(xc_n)+" "+str(xb_n)+" "+str(xa_n))
    #print(str(yd_n)+" "+str(yc_n)+" "+str(yb_n)+" "+str(ya_n))

    x_d.append(xd_n)
    y_d.append(yd_n)
    x_c.append(xc_n)
    y_c.append(yc_n)
    x_b.append(xb_n)
    y_b.append(yb_n)
    x_a.append(xa_n)
    y_a.append(ya_n)

    dist_dc= math.sqrt((x_d[t]-x_c[t])**2 + (y_d[t]-y_c[t])**2)
    dist_cb= math.sqrt((x_c[t]-x_b[t])**2 + (y_c[t]-y_b[t])**2)
    dist_ba= math.sqrt((x_b[t]-x_a[t])**2 + (y_b[t]-y_a[t])**2)
    dist_ad= math.sqrt((x_a[t]-x_d[t])**2 + (y_a[t]-y_d[t])**2)

    #print(str(dist_dc)+" "+str(dist_cb)+" "+str(dist_ba)+" "+str(dist_ad))
    if  dist_dc<10:
        count_dc=count_dc+1
        print("c_Fire_d "+ str(t))
    if  dist_cb<10:
        count_cb=count_cb+1
        print("b_Fire_c "+ str(t))
    if  dist_ba<10:
        count_ba=count_ba+1
        print("a_Fire_b "+ str(t))    
    if  dist_ad<10:
        count_ad=count_ad+1
        print("d_Fire_a "+ str(t))

print()
print("c shoots d "+str(count_dc)+" times")
print("d shoots c "+str(count_cb)+" times")
print("a shoots b "+str(count_ba)+" times")
print("d shoots a "+str(count_ad)+" times")
#print(x_d)
#print(y_d)   
print()
#print(x_c)
#print(y_c)   
print()
#print(x_b)
#print(y_b)   
print()
#print(x_a)
#print(y_a)   
print()
```
