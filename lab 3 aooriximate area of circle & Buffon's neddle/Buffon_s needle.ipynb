```
import numpy as np
import matplotlib.pyplot as plt

shots = [100,1000,5000,10000]
x = ['100','1000','5000','10000']
```

```
hits=[0,0,0,0]
difference=[0,0,0,0]
for i in range(4):
    pi=0
    
    for j in range (shots[i]):
        r = np.random.choice([0,1], p=[0.36,0.64])
        
        if r == 1:
            hits[i]=hits[i]+1
    
    print(hits[i])
    pi= 2*(shots[i]/hits[i])
    difference[i]=abs(3.1416-pi)
    print(pi)
    print(difference[i])
```

```

plt.bar(x, difference) 
plt.xlabel('number of trials')
plt.ylabel('error')
```
